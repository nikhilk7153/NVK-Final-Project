```python
def get_reviews_df(place_ids):
    reviews = []
    for business_id in place_ids:
        new_reviews = google_helper.get_reviews_from_id(business_id)
        reviews = reviews + new_reviews
        time.sleep(0.2)
    return pd.DataFrame(reviews, columns=['Review'])
```
